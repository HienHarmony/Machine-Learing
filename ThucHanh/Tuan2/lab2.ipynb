#Bài 1
Một giám đốc nhân sự mong muốn so sánh tính hiệu quả của hai phương pháp huấn luyện
các nhân viên công nghiệp nhằm thực hiện một hoạt động lắp ráp nào đó. Một số lượng
nhân viên được chia thành hai nhóm bằng nhau, n hóm thứ nhất nhận được phương pháp
huấn luyện 1 và nhóm thứ hai được huấn luyện bằng phương pháp 2. Mỗi nhóm sẽ thực
hiện hoạt động lắp ráp này, và độ dài của thời gian lắp ráp sẽ được ghi nhận. Người ta kỳ
vọng rằng các đại lượng cho cả hai nhóm sẽ có một khoảng xấp xỉ 8 phút. Để cho ước
lượng về sự khác biệt về thời gian trung bình để lắp ráp chính xác trong giới hạn 1 phút với
xác suất bằng với 0.95, thì cần phải đưa bao nhiêu công nhân vào mỗi nhóm huấn luyện?



 Cho $1.96 \sigma_{\left(\bar{x}_1-\bar{x}_2\right)}$ bằng với $B=1$ phút, chúng ta có được
$$
1.96 \sqrt{\frac{\sigma_1^2}{n_1}+\frac{\sigma_2^2}{n_2}}=1
$$

Bời vì chúng ta mong muốn $n_1$ bằng với $n_2$, chúng ta có thể đặt $n_1=n_2=n$ và có được phương trinh
$$
1.96 \sqrt{\frac{\sigma_1^2}{n}+\frac{\sigma_2^2}{n}}=1
$$

Như đã được lưu ý ở trên, độ biến thiên (dãy = range = max - min) của từng phương pháp lắp ráp xấp xỉ như nhau, và vì thế $\sigma_1^2=\sigma_2^2=\sigma^2$. Bởi vì dãy này, bằng với 8 phút, là xấp xi bằng với $4 \sigma$, cho nên
$$
4 \sigma \approx 8
$$
và
$$
\sigma \approx 2
$$

Thay thế giá trị này cho $\sigma_1$ và $\sigma_2$ trong phương trinh trên, chúng ta có được
$$
1.96 \sqrt{\frac{(2)^2}{n}+\frac{(2)^2}{n}}=1
$$
hay
$$
1.96 \sqrt{\frac{8}{n}}=1
$$
và
$$
\sqrt{n}=1.96 \sqrt{8}
$$

Giải ra, chúng ta có $n=31$. Như vậy mỗi nhóm nên có xấp xỉ $n=31$ công nhân.

```python
import math

def calculate_sample_size(sigma, margin_of_error, confidence_level):
    z_value = 1.96
    sample_size = math.ceil((z_value * sigma / margin_of_error)**2)

    return sample_size

sigma = 2
margin_of_error = 1
confidence_level = 0.95

sample_size = calculate_sample_size(sigma, margin_of_error, confidence_level)
print(f"Số lượng công nhân cần thiết cho mỗi nhóm: {sample_size}")
```

Biểu đồ cột có thể thể hiện so sánh giữa điểm khoa học (PV1SCIE) và điểm toán (PV1MATH) theo vùng địa lý (REGION). Điều này có thể giúp chúng ta nhận thức được sự khác biệt trong hiệu suất học tập giữa các khu vực khác nhau.


```python

def mean(data):
    return sum(data) / len(data)

def standard_deviation(data):
    mean_value = mean(data)
    squared_differences = [
        (value - mean_value) ** 2 for value in data
    ]
    return math.sqrt(sum(squared_differences) / len(data))

def range(data):
    return max(data) - min(data)

pv1MATH = data["PV1MATH"]
print("Mean:", mean(pv1MATH))
print("Standard deviation:", standard_deviation(pv1MATH))
print("Range:", range(pv1MATH))
```

#Bài 2
Một công ty kiểm toán mong muốn ước lượng sai số trung bình mỗi tài khoản trong các
khoản phải thu cho một công ty cung cấp hệ thống ống nước chính xác trong giới hạn 20 đô
với xác suất bằng 0.99. Một mẫu nhỏ trước đó gợi ý rằng sai số mỗi tài khoản sở hữu một
độ lệch chuẩn xấp xỉ bằng với 58 đô . Nếu công ty kiểm toán đó mong muốn ước lượng sai
số trung bình mỗi tài khoản chính xác trong giới hạn 20 đô, thì có bao nhiêu tài khoản ắt sẽ
phải được chọn mẫu? Mẫu này phải sở hữu (các) thuộc tính nào?

Công thức ước lượng kích thước mẫu (n) là:
$$
n=\left(\frac{Z^2 \times \sigma^2}{E^2}\right)
$$


$$
n=\left(\frac{(2.576)^2 \times(58)^2}{20^2}\right)
$$

Sau khi tính toán, chúng ta sẽ có giá trị của $n$, tức là kích thước mãu cân thiết để ước lượng sai số trung bình môi tài khoản trong giới hạn $\$ 20$ với độ tin cậy $99 \%$.

```python
import math

confidence_level = 0.99

z_score = 2.576

standard_deviation = 58

margin_of_error = 20

sample_size = math.ceil((z_score**2 * standard_deviation**2) / margin_of_error**2)

print(f"Kích thước mẫu cần chọn: {sample_size}")
```

```python
import numpy as np

units = np.array([3, 4, 1, 2, 5, 6, 4, 8, 9, 10, 10, 6, 7, 8, 9])
minutes = np.array([49, 64, 23, 29, 87, 96, 74, 145, 154, 166, 97, 109, 111, 149, 9])

def linear_regression(x, y):
    n = len(x)
    mean_x = np.mean(x)
    mean_y = np.mean(y)

    beta1 = np.sum((x - mean_x) * (y - mean_y)) / np.sum((x - mean_x)**2)

    beta0 = mean_y - beta1 * mean_x

    return beta0, beta1

beta0, beta1 = linear_regression(units, minutes)

print(f"β0 (hằng số): {beta0}")
print(f"β1 (hệ số của Units): {beta1}")

def predict_minutes(input_units, beta0, beta1):
    return beta0 + beta1 * input_units

try:
    input_units = float(input("Nhập số linh kiện cần sửa chữa: "))
    predicted_minutes = predict_minutes(input_units, beta0, beta1)
    print(f"Dự đoán số phút gọi vào tổng đài: {predicted_minutes:.2f} phút")
except ValueError:
    print("Vui lòng nhập một số linh kiện hợp lệ.")
```
